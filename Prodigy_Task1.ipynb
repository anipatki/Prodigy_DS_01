```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("/content/titanic_train.csv")
data.head(5)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=data,palette='winter')
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]

    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
data['Age'] = data[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
data.isnull().sum()
```

```python
sns.kdeplot(data['Age'], fill=True, color='skyblue', edgecolor='black')
```

The above graph shows the distribution of age among the passangers.After filling the missing values based on the average age according to the class we obtain the above density plot.
